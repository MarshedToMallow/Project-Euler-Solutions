# Factorial Digit Sum - Problem 20

$n!$ means $n \times (n-1)\times ...\times 3 \times 2 \times 1$.

For example, $10!=10\times 9\times ... \times 3 \times 2 \times 1 = 3628800$,

and the sum of the digits in the number $10!$ is $3+6+2+8+8+0+0=27$.

Find the sum of the digits in the number $100!$.

# My solution

Python's support for very large integers makes this fairly trivial.

We compute the factorial, convert it to a string, convert each digit back to integers, and sum the whole thing.

```python
from math import factorial

sum([int(i) for i in str(factorial(100))])
```
