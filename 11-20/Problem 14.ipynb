# Longest Collatz Sequence - Problem 14

The following iterative sequence is defined for the set of positive integers:

$n \rightarrow n/2$ ($n$ is even)
$n \rightarrow 3n+1$ ($n$ is odd)

Using the rule above and starting with 13, we generate the following sequence:

$13 \rightarrow 40 \rightarrow 20 \rightarrow 10 \rightarrow 5 \rightarrow 16 \rightarrow 8 \rightarrow 4 \rightarrow 2 \rightarrow 1$.

It can be seen that this sequence (starting with $13$ and finishing at $1$) contains $10$ terms.  Although it has not been proved yet (Collatz Problem), it is thought that all starting numbers finish at $1$.

Which starting number, under one million, produces the longest chain?

**NOTE:** Once the chain starts, the terms are allowed to go above one million.

# My solution

Collatz tends to be fairly quick for most starting values, without enough very long paths to hinder performance.

I have still included `lru_cache` anyway, but the speed improvement isn't crazy (12.5s to 6.5s).

Simply compute the path length for each value in the range and store the longest one so far.

```python
from functools import lru_cache

@lru_cache(10000)
def get_next(n):
    return n // 2 if n % 2 == 0 else 3 * n + 1

@lru_cache(10000)
def get_path_length(n):
    if n == 1:return 1
    return get_path_length(get_next(n)) + 1

best = 0
best_start = 0
for n in range(1, 10**6):
    length = get_path_length(n)
    if length > best:
        best = length
        best_start = n

best_start
```
