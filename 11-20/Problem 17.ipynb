# Number Letter Counts - Problem 17

If the numbers $1$ to $5$ are written out in words: one, two, three, four, five, then there are $3+3+5+4+4=19$ letters used in total.

If all the numbers from $1$ to $1000$ (one thousand) inclusive were written out in words, how many letters would be used?

**NOTE:** Do not count spaces or hyphens.  For example, 342 (three hundred and forty-two) contains $23$ letters and $115$ (one hundred and fifteen) contains $20$ letters.  The use of "and" when writing out numbers is in compliance with British usage.

# My solution

This one took me a bit.  I don't recall all the issues I had, but writing out the list of `assert` lines helps to catch the easy stuff.

First I defined a function to convert a number to its name.

Numbers below 20 are taken from a list of their names, which is also used at the end of other numbers (i.e. "twenty-six" uses "six").

Numbers below 100 have another list for the tens place and call the function on the ones place (unless its zero, for numbers in that range divisible by 10).

Numbers below 1000 work similarly where you have the hundreds place, which uses the same logic as the ones place, followed by the logic for the numbers below 100 if the overall number isn't divisible by 100.

1000 is hardcoded since there's only a single instance of a number that high.

After all that, we iterate over the values in the range and remove spaces and dashes from the names before adding the length to the total.

```python
SUB20 = "one two three four five six seven eight nine ten eleven twelve thirteen fourteen fifteen sixteen seventeen eighteen nineteen".split(" ")
TENS = "twenty thirty forty fifty sixty seventy eighty ninety".split(" ")

def number_to_name(n):
    if n < 20:return SUB20[n - 1]
    elif n < 100:
        tens = TENS[n // 10 - 2]
        if n % 10 == 0:
            return tens
        ones = number_to_name(n % 10)
        return f"{tens}-{ones}"
    elif n < 1000:
        hundreds = number_to_name(n // 100)
        if n % 100 == 0:
            return f"{hundreds} hundred"
        rest = number_to_name(n % 100)
        return f"{hundreds} hundred and {rest}"
    elif n == 1000:return "one thousand"
```

```python
assert sum([len(number_to_name(n)) for n in range(1, 6)]) == 19, "The problem says this should be 19"
assert number_to_name(342) == "three hundred and forty-two", f"The problem gave this example \"{number_to_name(342)}\" is wrong"
assert len(number_to_name(342).replace(" ","").replace("-", "")), f"The problem gave this example \"{number_to_name(342)}\" is wrong"
assert number_to_name(115) == "one hundred and fifteen", f"The problem gave this example \"{number_to_name(115)}\" is wrong"
assert len(number_to_name(115).replace(" ","").replace("-", "")), f"The problem gave this example \"{number_to_name(115)}\" is wrong"
```

```python
total_length = 0
for n in range(1, 1001):
    name = number_to_name(n)
    shortened_name = name.replace("-", "").replace(" ", "")
    total_length += len(shortened_name)

total_length
```
