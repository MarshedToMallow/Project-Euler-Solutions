# Lattice Paths - Problem 15

Starting in the top left corner of a $2 \times 2$ grid, and only being able to move right and down, there are exactly $6$ routes to the bottom right corner.

![](https://projecteuler.net/resources/images/0015.png?1678992052)

How many such routes are there through a $20 \times 20$ grid?

# My solution

This is a problem which lends itself to recursion.

At any given point, you can go down or you can go right.

So, we just take the sum of the paths for those two choices, recusively repeating that until one of the values hits a wall.

At that point, our options narrow down to just 1.

Sometimes we'll end up on a point we've visited before, so the `lru_cache` helps avoid the repeated computations.

```python
from functools import lru_cache

@lru_cache
def count_paths(w, h):
    if w == 0 or h == 0:return 1
    return count_paths(max(w-1, h), min(w-1, h)) + count_paths(max(w, h-1), min(w, h-1))

count_paths(20, 20)
```
