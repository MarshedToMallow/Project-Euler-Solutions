# Highly Divisible Triangular Number - Problem 12

The sequence of triangle numbers is generated by adding the natural numbers.  So the $7^{th}$ triangle number would be $1+2+3+4+5+6+7=28$.  The first ten terms would be:

$1,3,6,10,15,21,28,36,45,55,...$

Let us list the factors of the first seven triangle numbers:

```
1:  1
3:  1,3
6:  1,2,3,6
10: 1,2,5,10
15: 1,3,5,15
21: 1,3,7,21
28: 1,2,4,7,14,28
```

We can see that $28$ is the firs triangle number to have over five divisors.

What is the value of the first triangle number to have over five hundred divisors?

# My solution

A key insight here is that we don't actually need to know what all the divisors are, we just need the total number of them.

If we have, for example, $2^2\times 7^1=28$ then we can determine the number of divisors by looking at the possible exponents of its prime factors.

2 can have 0, 1, or 2 while 7 can have 0 or 1.  3 times 2, so 6.

So first, I define functions to handle prime factorization.

Then, we apply the counting divisors concept and simply iterate over the triangular numbers until we hit one with over 500 divisors.

````python
from functools import lru_cache

@lru_cache
def is_prime(n):
    if n == 1:return False
    if n == 2:return True
    if n % 2 == 0:return False
    for p in primes():
        if p * p > n:return True
        if n % p == 0:return False

prime_cache = [2, 3]
def primes():
    for p in prime_cache:yield p
    n = p + 2
    while True:
        if is_prime(n):
            prime_cache.append(n)
            yield n
        n += 2
````

````python
def count_divisors(n):
    total = 1
    for p in primes():
        if n == 1:return total
        count = 1
        while n % p == 0:
            count += 1
            n //= p
        total *= count

i = 1
while True:
    n = i * (i + 1) // 2
    count = count_divisors(n)
    if count > 500:
        break
    i += 1

n
````
