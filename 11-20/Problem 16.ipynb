# Power Digit Sum - Problem 16

$2^{15}=32768$ and the sum of its digits is $3+2+7+6+8=26$.

What is the sum of the digits of the number $2^{1000}$?

# My solution

Python supports very large integers, so this becomes as trivial as running the computation as described in the problem.

Below I've written it in two code blocks, one with easier-to-follow formatting and another as a single line of code.

```python
number = 2 ** 1000
number_string = str(number)
total = 0
for character in number_string:
    total += int(character)
total
```

```python
sum([int(i) for i in list(str(2**1000))])
```
