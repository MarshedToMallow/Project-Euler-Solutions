# Distinct Powers - Problem 29

Consider all integer combinations of $a^b$ for $2 \leq a \leq 5$ and $2 \leq b \leq 5$:

$2^2=4$, $2^3=8$, $2^4=16$, $2^5=32$

$3^2=9$, $3^3=27$, $3^4=81$, $3^5=243$

$4^2=16$, $4^3=64$, $4^4=256$, $4^5=1024$

$5^2=25$, $5^3=125$, $5^4=625$, $5^5=3125$

If they are then placed in numerical order, with any repeats removed, we get the following sequence of $15$ distinct terms:

$4, 8, 9, 16, 25, 27, 32, 64, 81, 125, 243, 625, 1024, 3125$.

How many distinct terms are in the sequence generated by $a^b$ for $2 \leq b \leq 100$ and $2 \leq b \leq 100$?

# My solution

Once again, Python's very large integers save the day.

Simply adding all the terms to a set and printing the length of that set gives us the correct answer.

```python
from itertools import product

terms = set()
for a, b in product(range(2, 101), repeat=2):
    terms.add(a ** b)

len(terms)
```
