# 1000-digit Fibonacci Number - Problem 25

The Fibonacci sequence is defined by the recurrence relation:

$F_n=F_{n-1}+F_{n-2}$, where $F_1=1$ and $F_2=1$.

Hence the first $12$ terms will be:

$F_1=1$

$F_2=1$

$F_3=2$

$F_4=3$

$F_5=5$

$F_6=8$

$F_7=13$

$F_8=21$

$F_9=34$

$F_{10}=55$

$F_{11}=89$

$F_{12}=144$

The $12th$ term, $F_{12}$ is the first term to contain three digits.

What is the index of the first term in the Fibonacci sequence to contain $1000$ digits?

# My solution

Python's large integer sizes come in handy yet again.

By performing a simple Fibonacci loop until the value has 1000 digits, we can easily obtain our result.

```python
a, b = 1, 1
index = 2

while True:
    index += 1
    a += b      # F_n = F_{n-1} + F_{n-2}
    b = a - b   # F_{n-1} = F_n - F_{n-2}
    if len(str(a)) >= 1000:
        break

print(index)
```
