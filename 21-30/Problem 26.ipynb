# Reciprocal Cycles - Problem 26

A unit fraction contains $1$ in the numerator.  The decimal representation of the unit fractions with denominators $2$ to $10$ are given:

$1/2=0.5$

$1/3=0.(3)$

$1/4=0.25$

$1/5=0.2$

$1/6=0.1(6)$

$1/7=0.(142857)$

$1/8=0.125$

$1/9=0.(1)$

$1/10=0.1$

Where $0.1(6)$ means $0.166666$ ..., and has a $1$-digit recurring cycle.  It can be seen that $1/7$ has a $6$-digit recurring cycle.

Find the value of $d<1000$ for which $1/d$ contains the longest recurring cycle in its decimal fraction part.

# My solution

This one was a bit tricky for me, but I managed to get a handle on it.

The idea is that our answer's digits eventually repeat, which is a consequence of seeing the same remainder (the part we subtract from in long division).

So, if we keep track of the remainders we've seen, then we can check if our current remainder is in there, returning the result based on that.

Keeping the repeating part separate from the non-repeating starting digits is reasonably trivial.

Then we just test the denominators below 1000 and keep track of which has the longest repeating portion.

```python
def reciprocal(n):
    if n == 1:return "1.0"

    digits = []
    remainders = [1]
    remainder = 10
    while remainder != 0:
        digits.append(remainder // n)
        remainder %= n

        if remainder in remainders:
            index = remainders.index(remainder)
            return "0." + "".join([str(i) for i in digits[:index]]) + "(" + "".join([str(i) for i in digits[index:]]) + ")"

        remainders.append(remainder)
        remainder *= 10
    return "0." + "".join([str(i) for i in digits])

longest = "0.()"
longest_denominator = 0
longest_length = 0
for n in range(2, 1000):
    r = reciprocal(n)
    if not "(" in r:continue

    if longest_length < len(r.split("(")[1][:-1]):
        longest = r
        longest_denominator = n
        longest_length = len(r.split("(")[1][:-1])

longest_denominator
```
