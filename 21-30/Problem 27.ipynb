# Quadratic Primes - Problem 27

Euler discovered the remarkable quadratic formula:

$n^2+n+41$

It turns out that the formula will produce $40$ primes for the consecutive integer values $0 \leq n \leq 39$. However, when $n=40$, $40^2+40+41=40(40+1)+41$ is divisible by $41$, and certainly when $n=41, 41^2+41+41$ is clearly divisible by $41$.

The incredible formula $n^2-79n+1601$ was discovered, which produces $80$ primes for the consecutive values $0 \leq n \leq 79$. The product of the coefficients, $-79$ and $1601$, is $-126479$.

Considering quadratics of the form:

> $n^2+an+b$, where $|a|<1000$ and $|b| \leq 1000$
> 
> where $|n|$ is the modulus/absolute value of $n$
> 
> e.g. $|11|=11$ and $|-4|=4$

Find the product of the coefficients, $a$ and $b$, for the quadratic expression that produces the maximum number of primes for consecutive values of $n$, starting with $n=0$.

# My solution

So doing around $2000^2$ checks isn't ideal, especially given we're checking primality each time.

That means we should do some analysis before iterating over potential answers.

First things first, we know that the case for $n=0$ is only prime if $b$ is prime, since $0^2+0\cdot a+b = b$

Similarly, we know that the case for $n=1$ is only prime if $1^2+1\cdot a+b \rightarrow a+b+1$ so we can check for values of $a$ based on what we know about values of $b$ (that they're prime).

From there, we can continue onto our final loop where we exhastively check every pair of values.

```python
from functools import lru_cache

@lru_cache(2**16)
def is_prime(n):
    if n < 2:return False
    for p in primes():
        if p * p > n:return True
        if n % p == 0:return False

def primes():
    yield 2
    n = 3
    while True:
        if is_prime(n):
            yield n
        n += 2
```

```python
b_values = []
for p in primes():
    if p > 1000:break
    b_values.append(p)

a_values = []
for i in range(-999, 1000):
    for b in b_values:
        if is_prime(i + b + 1):
            a_values.append(i)
            break

print(f"Originally we'd have had to check {1999 * 2001:,}")
print(f"Now it's only {len(a_values) * len(b_values):,}")
```

```python
from itertools import product

best = {
    "coefficients": {
        "a": None,
        "b": None
    },
    "list": [],
    "count": -1
}

for a, b in product(a_values, b_values):
    prime_list = [b]
    n = 1
    while True:
        result = n * n + a * n + b
        if not is_prime(result):break
        prime_list.append(result)
        n += 1
    if len(prime_list) > best["count"]:
        best = {
            "coefficients": {
                "a": a,
                "b": b
            },
            "list": prime_list,
            "count": len(prime_list)
        }
        print(best)

best["coefficients"]["a"] * best["coefficients"]["b"]
```
