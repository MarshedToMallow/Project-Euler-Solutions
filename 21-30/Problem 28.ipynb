# Number Spiral Diagonals - Problem 28

Starting with the number $1$ and moving to the right in a clockwise direction a $5$ by $5$ spiral is formed as follows:

```
21 22 23 24 25
20  7  8  9 10
19  6  1  2 11
18  5  4  3 12
17 16 15 14 13
```

It can be verified that the sum of the numers on the diagonals is $101$.

What is the sum of the numbers on the diagonals in a $1001$ by $1001$ spiral formed in the same way?

# My solution

Looking at the spiral, there's a pattern with how far separated each term is from the previous.

The first $4$ terms are $+2$, then the next $4$ are $+4$, and so on.

So we can implement a function which follows this pattern.

From there, we can figure the count of values to iterate with a simple formula, recognizing every time you increase the square by $2$, you get $4$ more numbers.

$2 * size - 1 = count$

````python
def diagonals():
    n = 1
    delta = 2
    while True:
        for i in range(4):
            yield n
            n += delta
        delta += 2

size = 1001
count = 2 * size - 1

total = 0
for i, _ in zip(diagonals(), range(count)):
    total += i

total
````
