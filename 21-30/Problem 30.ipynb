# Digit Fifth Power - Problem 30

Surprisingly there are only three numbers that can be written as a sum of fourth powers of their digits:

$1634 = 1^4 + 6^4 + 3^4 + 4^4$

$8208 = 8^4 + 2^4 + 0^4 + 8^4$

$9474 = 9^4 + 4^4 + 7^4 + 4^4$

As $1=1^4$ is not a sum it is not included.

The sum of these numbers is $1634+8208+9474=19316$.

Find the sum of all the numbers that can be written as the sum of fifth powers of their digits.

# My solution

So first, to avoid searching forever, we need a condition to end on.

The maximum digit is $9$ which means the maximum total sum for a given number of digits $d$ is $d \times 9^5$

This means that if $d \times 9^5 < 10^{d-1}-1$ then the digits can't possibly create a total large enough for their value.

From there, we can just keep a running total of the valid numbers until we hit that limit. 

```python
from itertools import product

running_total = 0
digit_count = 2

while (9 ** 5) * digit_count >= 10 ** (digit_count - 1) - 1:

    for digits in product(range(10), repeat=digit_count):
        if digits[0] == 0:continue

        total = sum([i ** 5 for i in digits])
        digit_string = "".join([str(i) for i in digits])

        if digit_string == str(total):
            running_total += total
            
    digit_count += 1

running_total
```
