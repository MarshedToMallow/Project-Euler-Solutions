# Summations of Primes - Problem 10

The sum of the primes below $10$ is $2+3+5+7=17$.

Find the sum of all the primes below two million.

# My solution

This is fairly straightforward, we just iterate over the primes until we hit our limit, being sure to add them to the total every iteration.

```python
def is_prime(n):
    if n == 1:return False
    if n == 2:return True
    if n % 2 == 0:return False
    for p in primes():
        if p * p > n:return True
        if n % p == 0:return False

prime_cache = [2, 3]
def primes():
    for p in prime_cache:yield p
    n = p + 2
    while True:
        if is_prime(n):
            prime_cache.append(n)
            yield n
        n += 2

total = 0
for p in primes():
    if p >= 2 * 10**6:break
    total += p

total
```
