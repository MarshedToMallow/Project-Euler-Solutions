# Smallest Multiple - Problem 5

2520 is the smallest number that can be divided by each of the numbers from 1 to 10 without any remainder.

What is the smallest positive number that is **evenly divisible** by all of the numbbers from 1 to 20?

# My solution

This problem is doable by hand if you're up to do it.

We look at each number in the range 1 to 20 and determine which prime factors exist and what their highest power is in that range.

Then, we simply multiply those prime factors raised to their highest exponents.

|Prime|Highest Power|
|---|---|
|2|2^4 (16)|
|3|3^2 (9)|
|5|5^1|
|7|7^1|
|11|11^1|
|13|13^1|
|17|17^1|
|19|19^1|

$2^4\cdot 3^2\cdot 5^1\cdot 7^1\cdot 11^1\cdot 13^1\cdot 17^1\cdot 19^1 = 232792560$

For the sake of completion, below is a program that does the same.

```python
def is_prime(n):
    if n == 1:return False
    if n == 2:return True
    if n % 2 == 0:return False
    for p in primes():
        if p * p > n:return True
        if n % p == 0:return False

prime_cache = [2, 3]
def primes():
    for p in prime_cache:yield p
    n = p + 2
    while True:
        if is_prime(n):yield n
        n += 2
```

```python
def get_prime_factors(n):
    factors = []
    for p in primes():
        if n == 1:return factors
        while n % p == 0:
            factors.append(p)
            n //= p
        if p in factors:
            if is_prime(n):return factors + [n]

powers = {}

for n in range(1, 21):
    factors = get_prime_factors(n)
    for f in factors:
        count = factors.count(f)
        if not f in powers or count > powers[f]:
            powers[f] = count

total = 1
for p in powers:total *= p ** powers[p]

total
```
