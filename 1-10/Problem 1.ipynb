# Multiples of 3 or 5 - Problem 1

If we list all the natural numbers below $10$ that are multiples of $3$ or $5$, we get $3, 5, 6$ and $9$.  The sum of these multiples is $23$.

Find the sum of all the multiples of $3$ or $5$ below $1000$.

# My solution

Using Python's `range()` and list comprehension, it's trivial to create a list of values divisible by some number below some limit.

By turning them into sets and getting their union, it easily removes duplicates, providing an easy input for `sum()` to give the final result.

```python
multiples_of_3 = set([n for n in range(3, 1000, 3)])
multiples_of_5 = set([n for n in range(5, 1000, 5)])

total = sum(multiples_of_3.union(multiples_of_5))
total
```
