# 10,001st Prime - Problem 7

By listing the first six prime numbers: $2, 3, 5, 7, 11$ and $13$, we can see that the $6th$ prime is $13$.

What is the $10001st$ prime number?

# My solution

It's as simple as generating every prime in order alongside every number up to 10,001

```python
def is_prime(n):
    if n == 1:return False
    if n == 2:return True
    if n % 2 == 0:return False
    for p in primes():
        if p * p > n:return True
        if n % p == 0:return False

prime_cache = [2, 3]
def primes():
    for p in prime_cache:yield p
    n = p + 2
    while True:
        if is_prime(n):
            prime_cache.append(n)
            yield n
        n += 2

for p, _ in zip(primes(), range(10001)):
    pass

p
```
