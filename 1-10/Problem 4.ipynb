# Largest Palindrome Product - Problem 4

A palindromic number reads the same both ways.  The largest palindrome made from the product of two $2$-digit numbers is $9009 = 91 \times 99$.

Find the largest palindrome made from the product of two $3$-digit numbers.

# My solution

I realize this problem can be solve trivially by simply creating a pair of for loops, but I wanted something different.

I was hoping to be able to iterate the pairs of factors from the largest product downward, but I'm not sure how to iterate with such a pattern beyond sorting a list already containing the values.

So instead of that, I opted to have a list with all of the multiples of 999.

Each iteration, I check if the largest value is a palindrome.  If it isn't, then I subtract out the value corresponding to the index.

As an example, if the largest value were at index 23, then that means the value is 24\*999 which means you can change it into 24\*998 by simply subtracting 1 more than the index.

In the event a value appears multiple positions, it just gets skipped every time after the first.

```python
products = [i * 999 for i in range(1, 1000)]
last_value = None

while True:
    value = max(products)
    if value != last_value:
        if str(value) == str(value)[::-1]:
            print(value)
            break
    last_value = value
    index = products.index(value)
    products[index] -= index + 1
```
