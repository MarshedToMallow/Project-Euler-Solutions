# Sum Square Difference - Problem 6

The sum of the squares of the first ten natural numbers is,

$1^2+2^2+...+10^2=385$

The square of the sum of the first ten natural numbers is,

$(1+2+...+10)^2=55^2=3025$

Hence the diference between the sum of the squares of the first ten natural numbers and the square of the sum is 3025 - 385 = 2640.

Find the difference between the sum of the squares of the first one hundred natural numbers and the square of the sum.

# My solution

Simply computing this directly in Python doesn't have any problems.

```python
def sum_squares(n):
    return sum([i*i for i in range(1, n+1)])

def square_sum(n):
    return  (n * (n + 1) // 2) ** 2

square_sum(100) - sum_squares(100)
```

# Another solution

I spent a bit looking at the pattern that both functions give for smaller values and realized they had fairly neat representations as simple operations on a couple of polynomials.

The sum of the first n values is already a reasonably trivial equation:

$\displaystyle\sum_{i=1}^{n}i=\frac{n\cdot (n+1)}{2}$

So we can simply square that for the squared sum:

$(\frac{n\cdot (n+1)}{2})^2$

Now the sum of squares is something I'm unfamiliar with, so I figured I'd start looking at the differences between consecutive terms.

Alright, so it seems to be a polynomial since the differences eventually become a constant.

Using sympy tends to make this sort of equation solving go much smoother, so let's try solving for this as a degree 3 polynomial.

```python
import sympy

a, b, c, d = sympy.var("a b c d")

(x1, y1), (x2, y2), (x3, y3), (x4, y4) = (1, 1), (2, 5), (3, 14), (4, 30)

equations = [
    sympy.Eq(a * x1 * x1 * x1 + b * x1 * x1 + c * x1 + d, y1),
    sympy.Eq(a * x2 * x2 * x2 + b * x2 * x2 + c * x2 + d, y2),
    sympy.Eq(a * x3 * x3 * x3 + b * x3 * x3 + c * x3 + d, y3),
    sympy.Eq(a * x4 * x4 * x4 + b * x4 * x4 + c * x4 + d, y4)
]

solution = sympy.solve(equations, a, b, c, d)

print(solution)
```

So we've found an equation which runs through those 4 points:

$f(x)=\frac{x^3}{3}+\frac{x^2}{2}+\frac{x}{6}$

From some quick research, this result is also well-known, though it is one I haven't committed to memory and working it out for myself was enjoyable.

It can also be rewritten as:

$\displaystyle\sum_{i=0}^{n}i^2=\frac{n\cdot (n + 1)\cdot (2n + 1)}{6}$

So now that we've done all that, we can turn this whole problem into a single algebraic expression, computed directly.

```python
def problem_6(n):
    return ((n * (n + 1) / 2) ** 2) - (n ** 3 / 3 + n ** 2 / 2 + n / 6)
```
