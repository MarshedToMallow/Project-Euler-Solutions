# Special Pythagorean Triplet - Problem 9

A Pythagorean triplet is a set of three natural numbers, $a < b < c$, for which,

$a^2+b^2=c^2$.

For example, $3^2+4^2=9+16=25=5^2$

There exists exactly one Pythagorean triplet for which $a + b + c = 1000$.

Find the product $abc$.

# My solution

I first created a `squares` function to generate pairs of natural numbers and their squares.

For the search, it's checking every possible pair of $a$ and $b$ values for an increasing $c$ value.

We have $a$ as the inner part of the search so that once it exceeds the computed value of $b$, we can skip to the next $c$, since $a < b < c$.

There's also a check to see if the $b^2$ created from the $b$ value matches up with the $b^2$ value from $a^2+b^2=c^2$.

If everything works out, then we've got our result.

```python
def squares():
    i = 1
    while True:
        yield (i, i * i)
        i += 1

done = False
for c, c2 in squares():
    for a, a2 in squares():

        if a >= c:break

        b = 1000 - a - c
        if a >= b:break

        b2 = b * b
        if b2 != c2 - a2:continue

        print(f"{a} * {b} * {c} = {a*b*c}")
        done = True
        break
    if done:break
```
