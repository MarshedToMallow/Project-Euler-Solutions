# Largest Prime Factor - Problem 3

The prime factors of $13195$ are $5, 7, 13$ and $29$.

What is the largest prime factor of the number $600851475143$?

# My solution

There are better approaches to generating primes, but this recursive approach has a sort of elegance I can't resist.

Simply start by yielding 2, then test each odd number for divisibility against the prime factors lower than it.

I've also included a list of cached primes, though for this particular problem, the performance isn't very significant.

</br>

From there, we can get the prime factors of a number by trying each of the primes.

If our value is divisible by the prime, we append it to the list of factors and divide it from the value.

Once the value hits 1, we've extracted all the possible prime factors.

The largest of those prime factors is our result.

```python
def is_prime(n):
    if n == 1:return False
    if n == 2:return True
    if n % 2 == 0:return False
    for p in primes():
        if p * p >= n:return True
        if n % p == 0:return False

prime_cache = [2, 3]
def primes():
    for p in prime_cache:yield p
    n = p + 2
    while True:
        if is_prime(n):yield n
        n += 2
```

```python
def get_prime_factors(n):
    factors = []
    for p in primes():
        if n == 1:return factors
        while n % p == 0:
            factors.append(p)
            n //= p
        if p in factors:
            if is_prime(n):return factors + [n]

max(get_prime_factors(600851475143))
```
