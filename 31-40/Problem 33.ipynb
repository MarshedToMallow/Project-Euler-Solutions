# Digit Cancelling Fractions - Problem 33

The fraction $49/98$ is a curious fraction, as an inexperienced mathematician in attempting to simplify it may incorrectly believe that $49/98=4/8$, which is correct, is obtained by cancelling the 9s.

We shall consider fractions like, $30/50=3/5$, to be trivial examples.

There are exactly four non-trivial examples of this type of fraction, less than one in value, and containing two digits in the numerator and denominator.

If the product of these four fractions is given in its lowest common terms, find the value of the denominator.

# My solution

Since it's relevant, I've imported the Fraction class from Python's built-in fractions module.

To avoid the trivial solutions mentioned, along with division by 0 errors and obviously incorrect numerators with 0s, I exclude values which are divisible by 10.

From there, we get the integer values of the numerator and denominator along with the set of the digit strings.

If the denominator is bigger, we can skip it, since we're only looking for solutions less than 1.

If they don't share any digits, then we can also skip it.

From there, we can test whether the operation maintains the same value, multiplying it to the total if so.

```python
from itertools import product
from fractions import Fraction



total = 1

for string_n, string_d in product([str(i) for i in range(10, 100) if i % 10 != 0], repeat=2):
    n, d = int(string_n), int(string_d)
    set_n, set_d = set(string_n), set(string_d)

    if d <= n:continue

    shared = set(string_n).intersection(set(string_d))
    if len(shared) == 0:continue

    f = Fraction(n, d)
    for s in shared:
        n2 = int(string_n.replace(s, "", 1))
        d2 = int(string_d.replace(s, "", 1))
        f2 = Fraction(n2, d2)
        if f2 == f:
            total *= f
            break

print(total)
```
