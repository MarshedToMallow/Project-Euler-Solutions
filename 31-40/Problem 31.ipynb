# Coin Sums - Problem 31

In the United Kingdom the currency is made up of pound (£) and pence (p).  There are eight coins in general circulation:

$1p, 2p, 5p, 10p, 20p, 50p, £1 (100p),$ and $£2 (200p)$.

It is possible to make $£2$ in the following way:

$1 \times £1 + 1 \times 50p + 2 \times 20p + 1 \times 5p + 1 \times 2p + 3 \times 1p$

How many different ways can $£2$ be made using any number of coins?

# My solution

A combination of dynamic programming and memoization can be applied here.

Our function tells us how many arrangements there are for a total of $n$.

We iterate over the possible coin values from smallest to largest.

If the current coin value is too big (i.e. bigger than the total OR bigger than the smallest coin used) we exit the loop and return the total.

Otherwise, we add the number of arrangements for $n$ minus the coin, passing the current coin as the new maximum.

```python
from functools import lru_cache

@lru_cache
def count_arrangements(n, biggest=200):
    if n == 0 or n == 1:return 1
    total = 0
    for i in [1, 2, 5, 10, 20, 50, 100, 200]:
        if i > biggest or i > n:break
        total += count_arrangements(n-i, i)
    return total

count_arrangements(200)
```
