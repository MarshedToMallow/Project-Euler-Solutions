# Circular Primes - Problem 35

The number, $197$, is called a circular prime because all rotations of the digits: $197, 971,$ and $719$, are themselves prime.

There are thriteen such primes below $100: 2, 3, 5, 7, 11, 13, 17, 31, 37, 71, 73, 79,$ and $97$.

How many circular primes are there below one million?

# My solution

First, we need to iterate over the primes.

With the `lru_cache` we're able to check primality once and have it saved for the more common values.

From there, we define `is_circular` with a custom cache.

If a value is found to be circular or not, that result is stored in the cache for that value and all its rotations.

From there, we can simply total count for the number of circular primes.

```python
from functools import lru_cache

@lru_cache(10000)
def is_prime(n):
    for p in primes():
        if p * p > n:return True
        if n % p == 0:return False

def primes():
    yield 2
    n = 3
    while True:
        if is_prime(n):
            yield n
        n += 2
```

```python
cache = {}
def is_circular(n):

    if n in cache:return cache[n]

    string_n = str(n)

    for i in range(1, len(string_n)):

        candidate = int(string_n[i:] + string_n[:i])

        if not is_prime(candidate):

            for j in range(len(string_n)):
                candidate = int(string_n[j:] + string_n[:j])
                cache[candidate] = False

            return False
        
    for j in range(len(string_n)):
        candidate = int(string_n[j:] + string_n[:j])
        cache[candidate] = True

    return True

circular_primes = 0

for p in primes():
    if p >= 10 ** 6:break
    if is_circular(p):
        circular_primes += 1

circular_primes
```
