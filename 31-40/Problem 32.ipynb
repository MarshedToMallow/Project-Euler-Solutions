# Pandigital Products - Problem 32

We shall say that an $n$-digit number is pandigital if it makes use of all the digits $1$ to $n$ exactly once; for example, the $5$-digit number, $15234$, is $1$ through $5$ pandigital.

The product $7254$ is unusual, as the identity, $39 \times 186=7254$, containing multiplicand, multiplier, and product is $1$ through $9$ pandigital.

Find the sum of all products whose multiplicand/multiplier/product identity can be written as a $1$ through $9$ pandigital.

HINT: Some products can be obtained in more than once way so be sure to only include it once in your sum.

# My solution

First, I created a function which helps decide which digit lengths are valid for $a$ and $c$ (the multiplicand and product respectively).

This follows these steps:

1. Return false if the $c$ is shorter than $a$
2. Return false if $b$ would be shorter than $a$
3. Return false if the smallest value for $a$ divided by the largest value of $c$ would require too many digits
4. Return true

This allows us to cut down on the search space significantly.

```python
from itertools import product

def are_valid_lengths(a_length, c_length):
    if a_length > c_length:return False

    b_length = 9 - a_length - c_length
    if b_length < a_length:return False

    smallest_a = int("1" + "0" * (1 - a_length))
    biggest_c = int("9" * c_length)

    if len(str(smallest_a // biggest_c)) > b_length:return False
    return True

# a * b = c
digit_lengths = [(a, c) for a, c in product(range(1, 10), repeat = 2) if are_valid_lengths(a, c)]
```

Next we have the main loop.

On the outermost loop, we iterate over the valid sizes of $a$ and $c$.

The next loop in, we iterate over tuples of digits for $c$ which we convert into the integer value and into a string.

We can then determine which digits can be used in $a$ by excluding ones in $c$.

Then another loop in we iterate over possible values of $a$, producing the integer value and string the same way.

We try various checks to prevent invalid pandigital identities:

1. Continue to next iteration if $c$ isn't divisible by $a$
2. Continue to next iteration if the size of $b$ doesn't match the amount leftover from $a$ and $c$
3. Continue to next iteration if there aren't 9 total digits in the combined $a, b,$ and $c$
4. Continue to next iteration if there's a 0 in $a, b,$ or $c$
5. If $c$ isn't in the product set, add it to the product set and then add $c$ to the total

From there, we get our final result.

```python
from itertools import permutations
from functools import reduce


total = 0
products = set()
for a_length, c_length in digit_lengths:


    for c_tuple in permutations(range(1, 10), r = c_length):


        c = int(reduce(lambda x, y: x * 10 + y, c_tuple))
        c_string = str(c)

        a_digits = set([i for i in range(1, 10) if not i in c_tuple])


        for a_tuple in permutations(a_digits, r = a_length):


            a = int(reduce(lambda x, y: x * 10 + y, a_tuple))
            a_string = str(a)

            if c % a != 0:continue

            b = c // a
            b_string = str(b)


            b_length = 9 - a_length - c_length
            if len(b_string) != b_length:continue

            if len(set(a_string + b_string + c_string)) != 9:continue

            if "0" in a_string + b_string + c_string:continue


            if not c in products:
                print(f"{a:,} * {b:,} = {c:,}")
                total += c
                products.add(c)
                
total
```
